```python
x=[1,2,3,4]
y=[3,3,3,3]
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

```python
x=np.arange(0,45,0.1)
s=np.sin(x)
c=np.cos(x)
plt.plot(x,s,label="sin",color='y')
plt.plot(x,c,label="cos",color='m')  
plt.legend()      
```

```python
dict={"Anu":100,"Anju":67,"Ram":75,"Ravi":89,"Manju":50,"Roy":65}
name=list(dict.keys())
mark=list(dict.values())
plt.bar(range(len(dict)), mark, tick_label=name,color='m')
plt.title("Student report")
plt.xlabel("Name")
plt.ylabel("Mark")
plt.show()
```
