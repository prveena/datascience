```python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
```

```python
from sklearn import tree
from sklearn.metrics import accuracy_score
```

```python
dt=tree.DecisionTreeClassifier()
dt.fit(X, y)
tree.plot_tree(dt)
y_pred = dt.predict(X)
print(y_pred)
print(accuracy_score(y, y_pred))
```
